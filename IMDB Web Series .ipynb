```python
import pandas as pd
from bs4 import BeautifulSoup
import requests

Name = []
Year = []
Runtime = []
Rating = []
Genre = []
Actors = []
Votes = []
count=0

url = "https://www.imdb.com/search/keyword/?keywords=web-series"

def get_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content,'html.parser')
    data = soup.findAll("div",{"class": "lister-item mode-detail"})

    for series in data:
        name=series.h3.a.text
        Name.append(name)
     
     
        
        year=series.h3.find("span", class_="lister-item-year text-muted unbold").text
        Year.append(year)

        run=series.find("span", class_="runtime").text.replace(" min",'') if series.find("span", class_="runtime") else None
        Runtime.append(run)

        rate=series.strong.text if series.strong else None
        Rating.append(rate)

        genre=series.find("span", class_="genre").text.strip() if series.find("span", class_="genre") else None
        Genre.append(genre)

        actor_con=series.find_all("p",class_="text-muted text-small")[1]
        actors=actor_con.find_all("a")
        a=[x.text for x in actors]
        act=" , ".join(a)
        Actors.append(act)
       


        vote = series.find("span",{"name":"nv"})["data-value"] if series.find("span",{"name":"nv"}) else None
        Votes.append(vote)
        
        
    try:
        url = soup.find("a", class_="lister-page-next next-page")["href"] 
    except:
        url=None
    if url==None:
        return None
    url = "https://www.imdb.com/search/keyword/"+url
    return url   

while 1:
    count+=1
    url = get_page(url)
    print(count)
    if url == None:
        break
        



movie_df=pd.DataFrame({"Name of the Movie": Name,"Year of Release": Year, "Runtime": Runtime , "Rating":Rating, "Genre": Genre,"Actors":Actors,"Votes":Votes})
movie_df.to_excel("Webseries_2.xlsx",index=False)

```
